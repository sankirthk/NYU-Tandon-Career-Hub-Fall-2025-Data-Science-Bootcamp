Q1. Define two custom numpy arrays, say A and B. Generate two new numpy arrays by stacking A and B vertically and horizontally.

```python
# Define two custom numpy arrays, say A and B. Generate two new numpy arrays by stacking A and B vertically and horizontally.
a = np.random.randint(low=1, high=20, size=(5, 1))
b = np.random.randint(low=1, high=20, size=(5, 1))
c = np.vstack((a, b))
d = np.hstack((a, b))
print("Vstacked Array", c)
print("HStacked Array", d)
```

Q2. Find common elements between A and B. [Hint : Intersection of two sets]

```python
# Find common elements between A and B. [Hint : Intersection of two sets]
e = np.intersect1d(a, b)
print("Intersection of A and B: ", e)
```

```python
# Extract all numbers from A which are within a specific range. eg between 5 and 10. [Hint: np.where() might be useful or boolean masks]
f = a[(a > 5) & (a < 15)]
print(f)
```

Q4. Filter the rows of iris_2d that has petallength (3rd column) > 1.5 and sepallength (1st column) < 5.0

```python
# Filter the rows of iris_2d that has petallength (3rd column) > 1.5 and sepallength (1st column) < 5.0

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

iris_2d = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[0,1,2,3])
iris_2d.size

filtered_data = iris_2d[(iris_2d[:, 2] > 1.5) & (iris_2d[:, 1] < 5.0 )]
filtered_data.size
```

Q3. How to get the rows of a dataframe with row sum > 100?

```python
# How to get the rows of a dataframe with row sum > 100?

df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
df.head()
df_rows_greater_hundred = df[df.sum(axis=1) > 100]
df_rows_greater_hundred
```
